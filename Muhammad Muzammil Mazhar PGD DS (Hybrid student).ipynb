```python
#question # 2 point 2 (total numbers of order place by each customer)

cst_wise_qty = {}


for i in data_py[1:]:
    if i[6] not in cst_wise_qty:
        cst_wise_qty[i[6]] = round(float(i[3]))
    else:
        cst_wise_qty[i[6]] += round(float(i[3]))
cst_wise_qty 


cst_wise_count = {}
for i in data_py[1:]:
    if i[6] not in cst_wise_count:
        cst_wise_count[i[6]] = 1
    else:
        cst_wise_count[i[6]] += 1

        

```

```python
#question # 2 point 3 (function to calculate average number of order placed each customer)

def avg_cust_sale(data_x,index_1,index_2,customerID):
    fun_cst_wise_total_sales = {}
    fun_cst_wise_count = {} 
    for i in data_x:
        if i[index_1] not in fun_cst_wise_total_sales:
            fun_cst_wise_total_sales[i[index_1]] = round(float(i[index_2]))
        else:
            fun_cst_wise_total_sales[i[index_1]] += round(float(i[index_2]))
    for i in data_x:
        if i[index_2] not in fun_cst_wise_count:
            fun_cst_wise_count[i[index_1]] = 1
        else:
            fun_cst_wise_count[i[index_1]] += 1
    return fun_cst_wise_total_sales[customerID]/fun_cst_wise_count[customerID]      

print(avg_cust_sale(data_py[1:],6,5,'17850')) # function to put data / index and customer ID
print(avg_cust_sale(data_py[1:],6,5,'14527'))
print(avg_cust_sale(data_py[1:],6,5,'15311'))
```

```python
#question # 2 point 4 (top 5 customer with highest average value)

total_sales_cust = {}


for i in data_py[1:]:
    if i[6] not in total_sales_cust:
        total_sales_cust[i[6]] = round(float(i[5]))
    else:
        total_sales_cust[i[6]] += round(float(i[5]))


count_cst_sale = {}
for i in data_py[1:]:
    if i[6] not in count_cst_sale:
        count_cst_sale[i[6]] = 1
    else:
        count_cst_sale[i[6]] += 1

cst_wise_avg_sale = {}        
for i in total_sales_cust:
    if i not in cst_wise_avg_sale:
        cst_wise_avg_sale[i] = round(total_sales_cust[i]/count_cst_sale[i])

cst_wise_avg_sale        
    
```

```python
#question # 3 part 1 (import liabraties and load data sheet)

import csv
import pandas as pd
import datetime as dt

whr_df = pd.read_csv('whr.csv',encoding = 'Latin - 1') 
print(whr_df.loc[0:5,])
print()
print()
whr_py = list(csv.reader(open('whr.csv',encoding='Latin - 1'))) # read data sets through python 

print(whr_py[0:5])
```

```python
#question # 3 part 2 (create new coloumn "happyness score category")

whr_py = list(csv.reader(open('whr.csv',encoding='Latin - 1')))
whr_py[0].append('Happiness Score Category')
for i in whr_py[1:]:
    if float(i[2]) < 5:
        i.append("Low")
    elif float(i[2]) <7:
        i.append("Medium")
    else:
        i.append("High")

print(whr_py[0:10])
```

```python
#question # 3 part 3 (create functrion to retuen country wise happiness")

def country_happy_type(data_y,index_contry,index_score,name):
    contry_score = {}
    for i in data_y[1:]:
        if i[index_contry] not in contry_score:
            contry_score[i[index_contry]] = i[index_score]
    return contry_score[name]        
              

print(country_happy_type(whr_py,1,-1,'Finland')) # function
print(country_happy_type(whr_py,1,-1,'Denmark')) # function
print(country_happy_type(whr_py,1,-1,'Pakistan'))
```
