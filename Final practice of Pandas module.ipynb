```python
def clean_date(date_x):
    bad_char = ["c","."," ","'","(",")","s"]
    for i in bad_char:
        date_x = date_x.replace(i,'')
    return date_x
```

```python
def clean_split(date_z):
    if '-' in date_z:
        date_split = date_z.split('-')
        return int((int(date_split[0])+int(date_split[1]))/2)
    else:
        return date_z
        
        
```
