```python
Iris_setosa = []
Iris_versicolor = []
Iris_virginica = []

for i in iris_py[1:]:
    flower = i[-1]
    petal_length = float(i[2])
    if flower == 'Iris-setosa':
        Iris_setosa.append(petal_length)
    elif flower == 'Iris-versicolor':
        Iris_versicolor.append(petal_length)
    elif flower == 'Iris-virginica':
        Iris_virginica.append(petal_length)

print('Iris-setosa',' ',sum(Iris_setosa)/len(Iris_setosa))
print('Iris-versicolor',' ',sum(Iris_versicolor)/len(Iris_versicolor))
print('Iris-virginica',' ',sum(Iris_virginica)/len(Iris_virginica))
```

```python
New_iris_py = [iris_py[0]]

for i in iris_py[1:]:
    petal_length = float(i[2])
    if petal_length > 5:
        New_iris_py.append(i)
        
print(New_iris_py)        
```

```python
sepal_lenght_list = []
sepal_width_list = []

for i in New_iris_py[1:]:
    sepal_lenght = float(i[0])
    sepal_width = float(i[1])
    sepal_lenght_list.append(sepal_lenght)
    sepal_width_list.append(sepal_width)
print('sepal-length-mean',' ',(sum(sepal_lenght_list)/len(sepal_lenght_list)))
print('sepal-width-mean',' ',(sum(sepal_width_list)/len(sepal_width_list))) 
```

```python
New_iris_df = iris_df[iris_df['petal_length'] > 5]
```

```python
New_iris_df['sepal_len_mean'] = New_iris_df['sepal_length'].sum()/New_iris_df['sepal_length'].count()
New_iris_df['sepal_width_mean'] = New_iris_df['sepal_width'].sum()/New_iris_df['sepal_width'].count()
```

```python
New_iris_df['sepal_len_sqdiff'] = (New_iris_df['sepal_length']-New_iris_df['sepal_len_mean'])**2
New_iris_df['sepal_wid_sqdiff'] = (New_iris_df['sepal_width']-New_iris_df['sepal_width_mean'])**2
```
